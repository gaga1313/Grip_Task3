```python
# state and profit
state_profit=pd.DataFrame({"State":data.State,"Profit":data.Profit})
```

```python
state_profit.head()
```

```python
## summing profit of each state
state_data=state_profit.groupby(["State"],as_index=False).sum()
state_data.head()
```

```python
state_data.shape
```

```python
##  there is a problem we have states with negative profit
colors = [plt.cm.Spectral(i/float(state_data.shape[0])) for i in range(state_data.shape[0])]
figure=plt.figure(figsize=(12,8))
ax=figure.add_subplot()
plt.bar(state_data["State"],state_data["Profit"],color=colors,edgecolor="black")

ax.tick_params(axis = 'x', labelrotation = 90, labelsize = 12)
ax.tick_params(axis = 'y', labelsize = 12)
ax.set_title("Profit_EAch_Sate", fontsize = 14);
```

```python
## lets try to explore sales and profit 
sales=pd.DataFrame({"State":data.State,"Sales":data.Sales,"Profit":data.Profit})
sales.head()
```

```python
# total sales and profit of each state
total_sales_profit=sales.groupby(["State"],as_index=False).sum()
total_sales=total_sales_profit.set_index("State")
total_sales_profit.head()
```

```python
## there are some states with high sales and low profit lets explore them and see how we can improve profit
fig = plt.figure(figsize = (16,8))
ax = fig.add_subplot()


ax.plot(total_sales["Sales"], color = "red", alpha = .5, label = "Sales")
ax.plot(total_sales["Profit"], color = "blue", alpha = .5, label = "Profit")

xtick_location = total_sales.index.tolist()
xtick_labels = total_sales.index.tolist()

ax.set_xticks(xtick_location)
ax.set_xticklabels(xtick_labels, rotation=90, fontdict={'horizontalalignment': 'center', 'verticalalignment': 'center_baseline'});


ax.legend(loc = "upper left", fontsize = 10)
ax.set_title("sales vs profit", fontsize = 14)
ax.grid( alpha = .7)
```

```python
df=total_sales_profit[total_sales_profit["Profit"]<0]
```

```python
df
```

```python
low_prf_state=data[data["State"].isin(df.State)]
```

```python
## count of sales of each sub-category
## seems binders,chairs,bookcasses etc are higher in count
plt.figure(figsize=(18,9))
sns.countplot(x = 'State', hue = 'Sub-Category', data = low_prf_state, palette= "rocket")
```

```python
sales=pd.DataFrame({"Sub-Category":low_prf_state["Sub-Category"],"Sales":low_prf_state.Sales,"Profit":low_prf_state.Profit})
sales_profit=sales.groupby("Sub-Category",as_index=True).sum()
sales_profit
```

```python
# we can stop selling items which have low sales like fasteners,Lablels
## seems these Low_prf_states donot have right buyers to buy these items
# though some category have high sales like binders , phones etc there profit is less ,lets see what is the problem
fig = plt.figure(figsize = (16,8))
ax = fig.add_subplot()


ax.plot(sales_profit["Sales"], color = "red", alpha = .5, label = "Sales")
ax.plot(sales_profit["Profit"], color = "blue", alpha = .5, label = "Profit")

xtick_location = sales_profit.index.tolist()
xtick_labels = sales_profit.index.tolist()

ax.set_xticks(xtick_location)
ax.set_xticklabels(xtick_labels, rotation=90, fontdict={'horizontalalignment': 'center', 'verticalalignment': 'center_baseline'});


ax.legend(loc = "upper left", fontsize = 10)
ax.set_title("sales vs profit", fontsize = 14)
ax.grid( alpha = .7)
```

```python
low_prf_state.head()
```

```python
## Binders have high sales but we are offering too much discount we need to reduce the discount ,we are going in loss.
#3 reducing the loss with help gain profit
#3 we can see current we are offering discount of 70% to 80% which is too much
#  similarly we can reduce the discount of other items to get profit
fig = plt.figure(figsize = (16,8))
ax = fig.add_subplot()

blinders=low_prf_state[low_prf_state["Sub-Category"]=="Binders"]
ax.plot(blinders["Discount"])
```

## Solutions thus far
### 1) In low profit states we can stop selling items with very low sales.
